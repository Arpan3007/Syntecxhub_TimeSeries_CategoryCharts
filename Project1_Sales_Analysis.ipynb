## üìå Section 4: Convert Date Columns 
We convert order_date and ship_date to datetime format and create Month and Quarter fields.


```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

df['Month'] = df['order_date'].dt.to_period('M')
df['Quarter'] = df['order_date'].dt.to_period('Q')

```

## üìå Section 4: CHARTS
## üìä Chart 1: Daily Sales Over Time (Line Chart)

### üîç Insight:
- The daily sales trend shows natural fluctuations across the timeline.
- Noticeable spikes occur during promotional periods and year-end months.
- Sales appear to increase toward the end of each year.
- This chart helps identify daily trends and anomalies.



```python
plt.figure(figsize=(12,6))
plt.plot(df['order_date'], df['sales'])
plt.title("Daily Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.savefig("outputs/daily_sales.png")
plt.show()
```

## üìä Chart 2: Monthly Sales Trend (Line Chart)

### üîç Insight:
- The monthly sales trend shows a clear upward movement from 2011 to 2014.
- Sales consistently rise toward the second half of each year, indicating strong seasonal demand.
- Noticeable peaks appear during mid-to-late year months, while early months (Jan‚ÄìFeb) show lower sales.
- This visualization reveals smooth long-term growth and highlights seasonal buying patterns.



```python
monthly_sales = df.groupby('Month')['sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.savefig("outputs/monthly_sales.png")
plt.show()
```

### üîç Insight:
- Quarterly sales show a clear upward trend from 2011 to 2014, indicating strong year-over-year business growth.
- Q3 and Q4 consistently generate higher sales across all years, with 2014 Q4 reaching the highest overall revenue.
- Early quarters (Q1 and Q2) show comparatively lower performance, highlighting seasonal demand that rises in the second half of each year.
- This pattern suggests that the business performs best during mid-to-late year periods, likely due to holiday seasons and increased consumer spending.



```python
quarterly_sales = df.groupby('Quarter')['sales'].sum()

plt.figure(figsize=(10,6))
quarterly_sales.plot(kind='bar')
plt.title("Quarterly Sales")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.savefig("outputs/quarterly_sales.png")
plt.show()
```

### üîç Insight:
- Technology records the highest total sales among all categories, indicating strong customer demand for tech-related products.
- Furniture contributes the second-highest sales, showing consistent performance but lower than Technology.
- Office Supplies generate the lowest revenue, likely due to lower-priced products and smaller order values.
- This comparison highlights the dominance of Technology in overall revenue contribution.


```python
category_sales = df.groupby('category')['sales'].sum()

plt.figure(figsize=(10,6))
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.savefig("outputs/category_sales.png")
plt.show()
```

## üìä Chart 5: Category-wise Sales Share (Pie Chart)

### üîç Insight:
- Technology holds the largest share of total sales, contributing approximately 37.5% of overall revenue.
- Furniture contributes around 32.5%, making it the second-largest segment.
- Office Supplies account for the remaining 30%, forming the smallest category by sales share.
- This distribution highlights the dominance of Technology products in driving total revenue.


```python
plt.figure(figsize=(8,8))
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Sales Share")
plt.ylabel("")
plt.savefig("outputs/category_pie.png")
plt.show()
```

# üé® Chart Choice and Formatting Discussion

In this project, different types of charts were used to effectively visualize sales patterns:

### 1. Line Charts (Daily & Monthly)
Line charts were chosen for daily and monthly sales because they clearly show changes over time.  
The continuous line helps identify trends, seasonal patterns, and fluctuations.  
Axis labels (‚ÄúDate‚Äù, ‚ÄúSales‚Äù, ‚ÄúMonth‚Äù, ‚ÄúTotal Sales‚Äù) were added to make the timeline and values easily readable.  
Gridlines were included to improve visual alignment of data points.

### 2. Bar Chart (Quarterly Sales)
A bar chart was used for quarterly sales because it makes comparison across quarters straightforward.  
Bars clearly highlight which quarters perform better.  
The x-axis shows quarters (e.g., 2011Q1), and the y-axis shows total sales for clarity.  
A simple bar format improves comparison of yearly performance.

### 3. Category-wise Bar Chart
Category sales were visualized using a bar chart since it is ideal for comparing numerical values across categories.  
Categories (Technology, Furniture, Office Supplies) are easy to compare side-by-side.  
Axis labels help interpret the values quickly.

### 4. Pie Chart (Category Sales Share)
A pie chart was used to show category share percentages in total sales.  
This format visually communicates proportion and contribution of each category.  
Percentage labels (autopct) help quantify share clearly.  
Pie charts are effective when categories are few and comparisons are proportional.

### Formatting Choices
- **Titles** were added to describe the purpose of each chart.
- **Axis labels** were included for clarity and interpretation.
- **Gridlines** were used in line charts to make reading values easier.
- **Consistent color themes** across bar and pie charts improve visual appeal.
- **Saving charts as PNG** ensures high-quality output for reports and GitHub submission.

Overall, these visualization and formatting choices were made to create clear, meaningful, and easy-to-interpret charts that effectively communicate patterns in the dataset.
